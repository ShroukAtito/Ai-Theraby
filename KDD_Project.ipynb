```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Perceptron
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

```

```python
data = pd.read_csv('Sample_KDD.csv')
data.head()
```

```python
X = data.drop('class', axis=1)
y = data['class']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

print("Training Set:")
print(y_train.value_counts())

print("\nTesting Set:")
print(y_test.value_counts())
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

perceptron = Perceptron(max_iter=1000, random_state=42)
perceptron.fit(X_train, y_train)
```

```python
y_pred = perceptron.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Perceptron Classification Accuracy =", accuracy)
```

```python
from sklearn.cluster import KMeans

for k in [2, 3, 4]:
    print(f"\nResults for K = {k}")

    kmeans = KMeans(n_clusters=k, random_state=42)
    clusters = kmeans.fit_predict(X_train)

    temp = X_train.copy()
    temp['cluster'] = clusters
    temp['class'] = y_train.values

    for c in range(k):
        print(f"\nCluster {c}:")
        print(
            temp[temp['cluster'] == c]['class']
            .value_counts(normalize=True) * 100
        )
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_train)
```

```python
clusters = kmeans.predict(X_train)

cluster_labels = {}

for i in range(4):
    cluster_labels[i] = y_train[clusters == i].value_counts().idxmax()

cluster_labels
```

```python
centroids = kmeans.cluster_centers_

predictions = []

for x in X_train.values:
    distances = np.linalg.norm(centroids - x, axis=1)
    closest_cluster = np.argmin(distances)
    predictions.append(cluster_labels[closest_cluster])
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_train, predictions)
print("Centroid-based Classification Accuracy =", accuracy)
```
